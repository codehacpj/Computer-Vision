### Idea behind GANs
GANs can create images on their own.<br>
Have 2 parts: <br>
<b>Generator</b> and <b>Discriminator</b><br>
these two components are constantly in touch, Generator generates images a nd the Discriminator then assesses those images and tells the generator if the images are similar to what actually exists or not.<br>
They start together, and they learn together. Generator generates some images, and then discriminator will look at the images that the generator generated and then it will look at the real table pictures, compare the two and then it learns and then it gives feedback to generator and thus next time the generator tries to come up with better image by itself.<br>

### How GANs work:
`G` stands for `generative`, represents the `Generator`: <br>
Which is a model which takes as input a random noise signal and then it outputs an image.<br>
<p>
    
`A` stands for `adversarial` and it represents the `Discriminator` which is the model which rivals `Generator` as it is capable of learning about objects or people or anything by learning the features.<br>
<p>

`N` stands for network and this represents the fact that both of above models are `Neural Networks`<br>

Working of GANs:<br>
<b>step 1</b><br>
input the noise signal into generator, then images are generated---> random images here, we have to train the discriminator for that we feed real images to discriminator<br>
all real images and created images are passed through discriminator since discriminator doesnot know anything about real and unreal it will assign random probabilities for different images to have detected features.<br>
Then it will be compared to the labels that it should have shown(the probabilities), and the error will be backpropagated to train the discriminator(i.e updating the weights on the network).<br>

Then to train the generator we take the same images that it created or regenerate the images and the we give it to discriminator again without real images.Then it gives some output. So this time as discrimator was trained, it outputs the lower probabilities for the random images created by the generator to have features of what it is looking for.<br>
Then the probabilities are compared to what it should have been(i.e should have had features as of real image so it is compared with `1`) and error is backpropagated to generator network to update its weights and thus next time it will do a better  job on creating images.<br>

## cifar-10 dataset is used now to generate the images using GANs.

torchvision is used to visualize the images. <br>
After Importing libraries and doing transforms and preprocessing ,load the dataset,initialize the weights of the given network.<br>

### Define the generator:
class for neural network---> sequence of different layers for the neural network, Then forward function to propagate the signal inside the neural network, Thus by doing this we can initialize the network.
`G` will be inherited from `nn.Module`.<br> `Meta module` called `Main` is made from nn.Sequential() class. The arguments to this will be the structure of neural network.<br>
The network contains the `inverse convolutions`---> `nn.ConvTranspose2d()` and this will take `100` input nodes, `512` feature maps, kernel of size `4`(`4*4`), `stride of 1`,`padding of 4` and the `bias is false`.(experimented values)<br>
`nn.BatchNorm2d()` used to normalize the feature maps.<br>
`Relu rectification`<br>
`Another inverse convolution` with the `512` feature maps as the input nodes and `256` feature maps<br>
Again batch norm the `256` feature maps.<br>
`Another inverse convolution`<br>
The output of generator has to to be fake images with 3 channels so the output of the generator has to be with 3 channels(i.e 3 feature maps).<br>
We end this up with `hyperbolic tangent` to break the linearity and to make sure the value is between 1 and -1(as we want the same standard as the images of the dataset as it will become the input for the `discriminaotr`)--->`nn.Tanh`.<br>
<p>
    
<b>signal forwarding functions.</b><br>
    It will take the random noise as the input in the first network of the Generator and forward it in the network.It takes the input passes it to the `Main` metamodule<br>
<p>
    
    
<b>Initialize the</b> `Generator`<br>
<b>Initialize the weights to</b> `Generator`   
    
### Define the Discriminator:
`Main meta module` is created as above but the modules inside it will be convolutions.<br>
It will take image from generator as input and will return the discriminating number(between 0 and 1), so  `Convolutions` will be added---> `nn.Conv2d`---> `3` as input size as the generator has `3` channels for outpu, `64` feature maps,`4` kernel size , `2` stride,`1` padding and `bias` is `False`.<br>
`LeakyRelu` is used for `rectification`, it is similar to Relu but besides having the max of zero and X, it has addition of negative slope multiplied by minimum of zeor and X. <br>
Redo this for next convolution, and here we keep on increasing the feature maps and then rectifying them using leakyrelu<br>
Redo this for next convolution, and here we keep on increasing the feature maps and then rectifying them using leakyrelu<br>
At last the output feature map will be `1` as will be a discriminating number<br>
Finally the `sigmoid activation` function used to get the values between 0 and 1.<br>
`Forward function` for the Discriminator, takes the input in the meta module main and returns the value(between 0 and 1).<br>
The result of the convolutions have to flattened using (`.view(-1)`).<br>
<b>Initialize Discriptor</b><br>
<b>Initialize weights for Discriptor</b><br>
<p>


### Train the networks:
update the weights of the discriminator networks, maximizing the error calculated by BCE(<b>Binary Cross Entropy</b>) loss and then updating weights in generator networks to maximize the error calculated by BCE loss.<br>
#### Training discriminator with real and fake images:
With the error calculated, we feed it to the `Generator` to apply stochastic gradient descent to update weights in generator.<br>
`Criterion` is defined for the images(the values between 0 and 1) and `Ground Truth`(always 0 or 1).<br>
The `Criterion` object will be instance of `BCEloss`<br>
`Optimizer` for Generator and `Optimizer` for the Discriminator.<br>

### Training with number of epochs:
loop over the dataloader:<br>
### step1: update the weights in the discriminator<br>
> Train discriminator with real data<br>
> Train discriminator with fake image generated by generator<br>
> Backpropagate the sum of errors from previous 2 steps.<br>

#### Training the discriminator with real image(ground truth):
1. Initialize the gradient of discriminant wrt the weight to zero.(`.zero_grad()`)<br>
2. Train the discriminator with real images from the minibatches from dataloader.<br> `data` has images and labels we take images from it.<br>
3. We convert the tensor above to the torch variable using `Variable()`. To make it input for the neural network.<br> 
4. The target here will be `1` as these are the real images so to create Variable of 1s we use `torch.zeros()` and pass `input.size()[0]` it will give number of real images<br>
5. Get the output passing input in netD<br>
6. error for real images is calculated by using criterion from above and passing output and target for it.<br>
#### Training the discriminator with fake image(ground truth for fake):
7. we do this with the noise images<br> Now to do that we use random vector which is made using `torch.randn()` takes arguments `batch size` mini batch of size of the input(`input.size()[0]`) Number of values in the vector will be 100 to create the mini batch of random vectors of size 100(as described in the first network of Generator it has input size of 100)`1,1` is given to add fake dimentions to the vector its like having 100 feature maps of size 1by1.<br> convert it to the Variable.<br>
8. create the fake images---> passing to netG.<br>
9. target here will be rejection(0s)---> tensor of zeros(torch.zeros(input.size[0],100,1,1).<br>
10. output will be netD(fake), to save up some memory as fake is a total variable because the output of Pytorch neural network is a torch variable so it contains tensor of predictions and discriminating numbers between 0 and 1 along with the gradients.<br>We don't need the gradients after backpropagation, so we can detach the gradient of this fake torche variable---> save up some memory and speed up the computations.<br>`fake.detach()`<br>
11. calculate the error from the `fake ground truth`--->give it to criterion with output and target. <br>
<p>

#### Backpropagating the total error in the discriminator.
12. Find total error<br>
13. Backpropagate the error<br>
14. apply the optimizer using---> optimizer.step(), step function applies the optimizer on the neural network of the discriminator to update the weights of the discriminator according to how much they are responsible for the concerning errors.<br>


### Step 2: Updating the weights of the neural network of the generator.

Initialize the gradients of the generator.<br>
The targets have to be vector of `1s` as we want to push the predictions to 1 so as to make the discriminator to accept that the fake images are real images.<br>
Get the output from discriminator for the fake images.<br>
get the error of predictions.<br>
This erro will be backpropagated to the generator.<br>
Then use the optimizer of generator to optimize the weights.<br>

### step 3: Print the loss and and saving the real images and generated images of the minibatch every 100 steps.


```python
# Deep Convolutional GANs

# Importing the libraries
from __future__ import print_function
import torch
import torch.nn as nn
import torch.nn.parallel
import torch.optim as optim
import torch.utils.data
import torchvision.datasets as dset
import torchvision.transforms as transforms
import torchvision.utils as vutils
from torch.autograd import Variable

# Setting some hyperparameters
batchSize = 64 
imageSize = 64 # size of the generated images (64x64).

# Creating the transformations
transform = transforms.Compose([transforms.Scale(imageSize), transforms.ToTensor(), transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5)),]) # We create a list of transformations (scaling, tensor conversion, normalization) to apply to the input images.

# Loading the dataset
dataset = dset.CIFAR10(root = './data', download = True, transform = transform) 
dataloader = torch.utils.data.DataLoader(dataset, batch_size = batchSize, shuffle = True, num_workers = 2) 

# initialize all its weights.
def weights_init(m):
    classname = m.__class__.__name__
    if classname.find('Conv') != -1:
        m.weight.data.normal_(0.0, 0.02)
    elif classname.find('BatchNorm') != -1:
        m.weight.data.normal_(1.0, 0.02)
        m.bias.data.fill_(0)

# Defining the generator

class G(nn.Module):

    def __init__(self):
        super(G, self).__init__()
        self.main = nn.Sequential(
            nn.ConvTranspose2d(100, 512, 4, 1, 0, bias = False),
            nn.BatchNorm2d(512),
            nn.ReLU(True),
            nn.ConvTranspose2d(512, 256, 4, 2, 1, bias = False),
            nn.BatchNorm2d(256),
            nn.ReLU(True),
            nn.ConvTranspose2d(256, 128, 4, 2, 1, bias = False),
            nn.BatchNorm2d(128),
            nn.ReLU(True),
            nn.ConvTranspose2d(128, 64, 4, 2, 1, bias = False),
            nn.BatchNorm2d(64),
            nn.ReLU(True),
            nn.ConvTranspose2d(64, 3, 4, 2, 1, bias = False),
            nn.Tanh()
        )

    def forward(self, input):
        output = self.main(input)
        return output

# Creating the generator
netG = G()
netG.apply(weights_init)

# Defining the discriminator

class D(nn.Module):

    def __init__(self):
        super(D, self).__init__()
        self.main = nn.Sequential(
            nn.Conv2d(3, 64, 4, 2, 1, bias = False),
            nn.LeakyReLU(0.2, inplace = True),
            nn.Conv2d(64, 128, 4, 2, 1, bias = False),
            nn.BatchNorm2d(128),
            nn.LeakyReLU(0.2, inplace = True),
            nn.Conv2d(128, 256, 4, 2, 1, bias = False),
            nn.BatchNorm2d(256),
            nn.LeakyReLU(0.2, inplace = True),
            nn.Conv2d(256, 512, 4, 2, 1, bias = False),
            nn.BatchNorm2d(512),
            nn.LeakyReLU(0.2, inplace = True),
            nn.Conv2d(512, 1, 4, 1, 0, bias = False),
            nn.Sigmoid()
        )

    def forward(self, input):
        output = self.main(input)
        return output.view(-1)

# Creating the discriminator
netD = D()
netD.apply(weights_init)

# Training the DCGANs

criterion = nn.BCELoss()
optimizerD = optim.Adam(netD.parameters(), lr = 0.0002, betas = (0.5, 0.999))
optimizerG = optim.Adam(netG.parameters(), lr = 0.0002, betas = (0.5, 0.999))

for epoch in range(25):

    for i, data in enumerate(dataloader, 0):
        
        # Step1: Updating the weights of the neural network of the discriminator

        netD.zero_grad()
        
        # Training the discriminator with a real image of the dataset
        real, _ = data
        input = Variable(real)
        target = Variable(torch.ones(input.size()[0]))
        output = netD(input)
        errD_real = criterion(output, target)
        
        # Training the discriminator with a fake image generated by the generator
        noise = Variable(torch.randn(input.size()[0], 100, 1, 1))
        fake = netG(noise)
        target = Variable(torch.zeros(input.size()[0]))
        output = netD(fake.detach())
        errD_fake = criterion(output, target)
        
        # Backpropagating the total error
        errD = errD_real + errD_fake
        errD.backward()
        optimizerD.step()

        # 2nd Step: Updating the weights of the neural network of the generator

        netG.zero_grad()
        target = Variable(torch.ones(input.size()[0]))
        output = netD(fake)
        errG = criterion(output, target)
        errG.backward()
        optimizerG.step()
        
        # 3rd Step: Printing the losses and saving the real images and the generated images for every 100 iterations(minibatch)

        print('[%d/%d][%d/%d] Loss_D: %.4f Loss_G: %.4f' % (epoch, 25, i, len(dataloader), errD.data[0], errG.data[0]))
        if i % 100 == 0:
            vutils.save_image(real, '%s/real_samples.png' % "./results", normalize = True)
            fake = netG(noise)
            vutils.save_image(fake.data, '%s/fake_samples_epoch_%03d.png' % ("./results", epoch), normalize = True)
```
